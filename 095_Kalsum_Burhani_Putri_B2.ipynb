####Outliers Values

```
def detect_outliers(df):
    outliers_dict = {}
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
            outliers_count = outliers.shape[0]
            outliers_dict[column] = outliers_count
    return outliers_dict

outliers_result = detect_outliers(df)
print("Outliers count in each column:")
print(outliers_result)
```



1. Pada Index: Tidak ada outliers dalam kolom ini, dengan jumlah outliers sebesar 0. Ini menunjukkan bahwa semua nilai dalam kolom indeks berada dalam rentang yang dianggap normal.
2. Pada Selling Price: Terdapat 19 outliers dalam kolom harga jual. Outliers dalam konteks ini adalah nilai-nilai yang jauh dari rentang nilai yang dianggap normal, mungkin merupakan harga yang sangat tinggi atau rendah dibandingkan dengan harga rata-rata produk.
3. Pada Average Rating: Terdapat 35 outliers dalam kolom rating rata-rata. Outliers dalam konteks ini bisa berarti produk-produk yang mendapatkan penilaian yang sangat tinggi atau rendah dibandingkan dengan rata-rata umum.
4. Pada Reviews Count: Terdapat 93 outliers dalam kolom jumlah ulasan. Outliers dalam konteks ini mungkin menunjukkan produk-produk yang menerima jumlah ulasan yang sangat tinggi atau rendah dibandingkan dengan produk lainnya.



####Duplicated Values

```
def check_duplicates(df):
    duplicates_dict = {}
    for column in df.columns:
        duplicates_count = df.duplicated(subset=[column]).sum()
        duplicates_dict[column] = duplicates_count
    return duplicates_dict

duplicates_result = check_duplicates(df)
print("Duplikat count in each column:")
print(duplicates_result)
```
